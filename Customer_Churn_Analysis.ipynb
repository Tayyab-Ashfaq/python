```python
import plotly.express as px
```

```python

df = pd.read_csv('IBM Customer Churn.csv')
```

```python
df.head()
```

```python
# Scatter plot between two numerical feature
fig = px.scatter(df, x='MonthlyCharges', y='TotalCharges', color='Churn', title='Monthly vs Total Charges by Churn')
fig.show()
```

```python
# Histogram of customer tenure
fig = px.histogram(df, x='tenure', nbins=30, title='Distribution of Customer Tenure')
fig.show()
```

```python
# Line chart of MonthlyCharges over tenure
fig = px.line(df, x='tenure', y='MonthlyCharges', title='Monthly Charges over Tenure')
fig.show()
```

```python
plt.figure(figsize = (5,3))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.pie(gb['Churn'], labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churned Customeres", fontsize = 10)
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.ticker as mtick
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
telco_base_data = pd.read_csv('/content/Churn_Modelling.csv')
```

```python
telco_base_data.head(5)
```

## Age vs Tenure

```python
# @title Age vs Tenure

from matplotlib import pyplot as plt
telco_base_data.plot(kind='scatter', x='Age', y='Tenure', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
telco_new = telco_base_data[['Geography','Gender','Exited']]
```

```python
telco_new.head(5)
```

```python
# @title Geography vs Gender

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['Gender'].value_counts()
    for x_label, grp in telco_new.groupby('Geography')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('Geography')
_ = plt.ylabel('Gender')
```
